### Test functions
This cell defines 4 functions:
- `test_interpolation_fn(fn, samples)`
- `set_metric(name, fn)`
- `del_metric(name)`
- `linear_interpolation(decompose, recompose)`

The main test method is `test_interpolation_fn(fn, samples)`. It accepts:
- `fn(first, last, n)` - A function that generates interpolated
matrices. `first` is the first matrix in the sequence, `last` is the
last matrix in the sequence, and `n` is the total number of frames
required. This is the function we need to discover. We need to
figure out how Adobe Animate generates in-between matrices given
the starting point, ending point, and number of frames.
- `samples` - A sequence of frames to test the function on. You can
get these from the `data` variable. `samples=data['scale_only'][0]`
is an easy one to get started on. The hardest one is likely going to
be `samples=data['full'][1]`.

`test_interpolation_fn` generates a bunch of metrics and plots the
results. The list of metrics is stored in the `metrics` variable.
The `metrics` variable is by default populated by functions that
calculate the relative error element-by-element between the ground
truth matrices and the generated matrices. You can add more metrics
using `set_metric(name, fn)`:
- `name` - Each metric will be plotted on a graph using pyplot. The
graph will be titled by this name. This name should be distinct for
each metric. You can overwrite an existing metrics using `set_metric`
with a `name` that's already in use.
- `fn` - This function should accept whatever parameters it requires
and output a single scalar value for each invocation. The following
named arguments are passed to `fn` on each invocation: `truth`
(ground truth), `generated` (the interpolated matrix to check),
`additive_diff` (the element-by-element difference from the ground
truth to the interpolated matrix), `percent_diff` (the
element-by-element difference scaled by the ground truth value),
`linear_diff` (the matrix transformation required to turn the ground
truth to the generated value).

You can remove metrics using the `del_metric(name)` function.

The `linear_interpolation(decompose, recompose)` function is a helper
function to generate interpolated matrices. It does the following:
- Use the `decompose` function to turn start and end transformation
matrices into a set of parameters.
- Linearly interpolate those parameters as needed.
- Use the `recompose` function to turn interpolated parameters into
transformation matrices.
- Wrap the above functionality into a function that can be passed to
`test_interpolation_fn`.

At a high level, it's used like this:
> ```
def matrix_to_parameters(matrix):
  """ Convert a matrix into some parameters. """"
def parameters_to_matrix(*params):
  """ Convert parameters into a matrix. """
fn = linear_interpolation(matrix_to_parameters, parameters_to_matrix)
test_interpolation_fn(fn, data['full'][0])
 ```

````python
def linear_interpolation(decompose, recompose):
    def _gen_interpolated_values(start, end, n_frames):
        f1_decomp = decompose(start)
        fn_decomp = decompose(end)
        
        for i in range(n_frames):
            frac = i / (n_frames - 1)
            interpolated_values = [
                frac*fn_d + (1-frac)*f1_d
                    for fn_d, f1_d
                    in zip(fn_decomp, f1_decomp)
            ]
            
            yield recompose(*interpolated_values)
    
    return _gen_interpolated_values

def _sb_plot(figure, title, i, value):
    plt.figure(figure)
    plt.title(title)
    plt.plot(i, value, 'k.')

metrics = {}

def _sb_check_interpolation(i, truth, interp):
    additive_diff = truth - interp
    kwargs = {
            'truth': truth,
            'generated': interp,
            'additive_diff': additive_diff,
            'percent_diff': additive_diff / (truth + 1e-10),
            'linear_diff': numpy.linalg.inv(truth) @ interp,
    }

    for plot_idx, key in enumerate(metrics.keys()):
        metric_fn = metrics[key]
        _sb_plot(plot_idx, key, i, metric_fn(**kwargs))


def test_interpolation_fn(fn, samples):
    n_frames = len(samples)
    interpolations = list(fn(samples[0], samples[-1], n_frames))
    for i, truth, interp in zip(range(n_frames), samples, interpolations):
        _sb_check_interpolation(i, truth, interp)
    
    for i in range(len(metrics)):
        plt.figure(i)
        plt.show()
        plt.close()

def set_metric(name, fn):
    metrics[name] = fn

def del_metric(name):
    del metrics[name]

set_metric('matrix error[0,0]', lambda additive_diff, **kwargs: additive_diff[0,0])
set_metric('matrix error[0,1]', lambda additive_diff, **kwargs: additive_diff[0,1])
set_metric('matrix error[1,0]', lambda additive_diff, **kwargs: additive_diff[1,0])
set_metric('matrix error[1,1]', lambda additive_diff, **kwargs: additive_diff[1,1])
````
